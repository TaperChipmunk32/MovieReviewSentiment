```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
def runModels(data, corpus):
    vectorizer = TfidfVectorizer()
    X = vectorizer.fit_transform(corpus).toarray()
    y = data['category'].values
    kf = KFold(n_splits=10)
    kf.get_n_splits(X)
    NB_accuracies = []
    DT_accuracies = []
    i = 1
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        NB_accuracy = NBRun(X_train, y_train, X_test, y_test)
        NB_accuracies.append(NB_accuracy)
        print(f'NB Run {i} Accuracy: {NB_accuracy}')
        DT_accuracy = DTRun(X_train, y_train, X_test, y_test)
        DT_accuracies.append(DT_accuracy)
        print(f'DT Run {i} Accuracy: {DT_accuracy}')
        i = i + 1
    return NB_accuracies, DT_accuracies
```
